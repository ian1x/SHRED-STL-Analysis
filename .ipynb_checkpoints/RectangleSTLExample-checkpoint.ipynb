```python
#building a cube in order to test eigenvectors and translations/rotations
from stl import mesh
from mpl_toolkits import mplot3d
from matplotlib import pyplot
import numpy as np
import matplotlib.style as mplstyle
```

```python
vertices = np.array([[0,0,0], #0
            [1,0,0], #1
            [1,2,0], #2
            [0,2,0], #3
            [0,0,1], #4
            [1,0,1], #5
            [1,2,1], #6
            [0,2,1]]) #7
faces = np.array([[0,5,4],
         [4,0,5],
         [5,0,1],
         [6,5,1],
         [2,6,1],
         [7,6,3],
         [3,6,2],
         [0,4,7],
         [0,7,3],
         [4,5,7],
         [5,6,7],
         [0,3,1],
         [1,3,2]])
rectprism = mesh.Mesh(np.zeros(faces.shape[0], dtype=mesh.Mesh.dtype))
for i, f in enumerate(faces):
    for j in range(3):
        print(vertices[f[j],:])
        rectprism.vectors[i][j] = vertices[f[j]]

# Write the mesh to file "cube.stl"
rectprism.save('rectprism.stl')
```

```python
%matplotlib widget
figure = pyplot.figure()
axes = figure.add_subplot(projection='3d')

# Load the STL files and add the vectors to the plot
axes.add_collection3d(mplot3d.art3d.Poly3DCollection(rectprism.vectors))

# Auto scale to the mesh size
scale = rectprism.points.flatten()
axes.auto_scale_xyz(scale, scale, scale)

# Show the plot to the screen
pyplot.show()
```
